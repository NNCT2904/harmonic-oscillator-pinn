```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv('plots/PINN_113.csv', index_col=0)
df2 = pd.read_csv('plots/PINN_2209.csv', index_col=0)
df3 = pd.read_csv('plots/PIQML_109.csv', index_col=0)
```

```python
df1['0'].iloc[-1]
```

```python
plt.figure(figsize=(8,4))
plt.title("Estimation of $\mu$ with training of Physics Informed Models")
plt.hlines(4, 0, len(df1), label="True value = 4", colors="tab:red")
plt.plot(df1, label=f"PINN_113, ~ {round(df1['0'].iloc[-1], 2)}")
plt.plot(df2, label=f"PINN_2209, ~ {round(df2['0'].iloc[-1], 2)}")
plt.plot(df3, label=f"PIQML_109, ~ {round(df3['0'].iloc[-1], 2)}")
plt.legend(loc='best')
plt.show()
```
